### FOR LONG RUN (10hrs)

```python
s=time()
data=data.sort_values(by=['client', 'doc_date'])
data.reset_index(drop=True, inplace=True)
data['rownum']=data.index
clients = list(pd.unique(data.client))
# data.head()
sort_time = (time()-s)/60
print("sort time:", sort_time)

print(len(clients))

s=time()

clients_sample = clients[:20]

client_features = []

# for client in clients_sample:
for client in clients:
    client_ = data[data.client==client]
    print(client)
    
    for i in range(min(client_.rownum), max(client_.rownum)):
        if client_.calculate_additional_features_flag[i] == 'N':
            continue
        else:
#         print(i)
            yesterday = client_.doc_date[i] - pd.DateOffset(days = 1)
            client_sub = client_[(client_.doc_date<=yesterday)]
    #         print(len(client_sub))
            try:
                client_yrs = (max(client_sub.doc_date) - min(client_sub.doc_date)).year +1
            except:
                client_yrs = 1
            client_past_inv_cnt = len(client_sub)
            client_annual_inv_cnt = client_past_inv_cnt*1./client_yrs*1.
            client_ave_fin_amt = np.mean(client_sub.fin_usd_amt)
            client_ave_annual_fin_amt = np.sum(client_sub.fin_usd_amt) / client_yrs*1.
            client_std_fin_amt = np.std(client_sub.fin_usd_amt)
            client_late_days = np.mean(client_sub.late_days)
            client_isLate = np.mean(client_sub.isLate)
            client_dr = np.mean(client_sub.dr)
            client_term = np.mean(client_sub.fin_tenor)
            client_cty = len(pd.unique(client_sub.cty_code))

    #         print(client_past_inv_cnt)
            prev_yr = str(client_.doc_date[i] - pd.DateOffset(years = 1))
            client_prevyr_sub = client_sub[
        #         (data.doc_date < data.doc_date[i]) 
                (client_sub.rownum < i) 
                &(client_sub.doc_date >=prev_yr)
        #         & (data.client == data.client[i] )
                ]
            client_prevyr_inv_cnt = len(client_prevyr_sub)
            client_prevyr_ave_fin_amt = np.mean(client_prevyr_sub.fin_usd_amt)
            client_season = client_.fin_usd_amt[i]/client_prevyr_ave_fin_amt
            client_prevyr_std_fin_amt = np.std(client_prevyr_sub.fin_usd_amt)
            client_prevyr_late_days = np.mean(client_prevyr_sub.late_days)
            client_prevyr_isLate = np.mean(client_prevyr_sub.isLate)
            client_prevyr_dr = np.mean(client_prevyr_sub.dr)
            client_prevyr_term = np.mean(client_prevyr_sub.fin_tenor)
        #     client_prevyr_cty = len(pd.unique(client_prevyr_sub.cty_code))


            prev_6m = str(client_.doc_date[i] - pd.DateOffset(months = 6))
            client_6m_sub = client_prevyr_sub[
        #         (data.doc_date < data.doc_date[i]) 
                (client_prevyr_sub.rownum < i) 
                &(client_prevyr_sub.doc_date >= prev_6m)
        #         & (data.client == data.client[i] )
                ]

            client_6m_inv_cnt = len(client_6m_sub)
            client_6m_ave_fin_amt = np.mean(client_6m_sub.fin_usd_amt)
            client_6m_std_fin_amt = np.std(client_6m_sub.fin_usd_amt)
            client_6m_dr = np.mean(client_6m_sub.dr)
            client_6m_term = np.mean(client_6m_sub.fin_tenor)
        #     client_6m_cty = len(pd.unique(client_6m_sub.cty_code))
            client_6m_late_days = np.mean(client_6m_sub.late_days)
            client_6m_isLate = np.mean(client_6m_sub.isLate)

            prev_3m = str(client_.doc_date[i] - pd.DateOffset(months = 3))
            client_3m_sub = client_prevyr_sub[
        #         (data.doc_date < data.doc_date[i]) 
                (client_prevyr_sub.rownum < i) 
                &(client_prevyr_sub.doc_date >= prev_3m)
        #         & (data.client == data.client[i] )
                ]

            client_3m_inv_cnt = len(client_3m_sub)
            client_3m_ave_fin_amt = np.mean(client_3m_sub.fin_usd_amt)
            client_3m_std_fin_amt = np.std(client_3m_sub.fin_usd_amt)
            client_3m_dr = np.mean(client_3m_sub.dr)
            client_3m_term = np.mean(client_3m_sub.fin_tenor)
        #     client_3m_cty = len(pd.unique(client_6m_sub.cty_code))
            client_3m_late_days = np.mean(client_3m_sub.late_days)
            client_3m_isLate = np.mean(client_3m_sub.isLate)

            client_features.append([i,data.doc_id[i], client_yrs, client_past_inv_cnt, client_annual_inv_cnt, client_ave_fin_amt, 
                                client_ave_annual_fin_amt, client_std_fin_amt, client_late_days, client_isLate, client_dr, 
                                client_term, client_cty, client_prevyr_inv_cnt, 
                                client_prevyr_ave_fin_amt, client_prevyr_std_fin_amt, client_prevyr_late_days, 
                                client_prevyr_isLate, client_prevyr_dr, client_prevyr_term,  
                                client_6m_inv_cnt, 
                                client_6m_ave_fin_amt, client_6m_std_fin_amt, client_6m_dr, client_6m_term, 
                                client_6m_late_days, client_6m_isLate, 
                                client_3m_inv_cnt, client_3m_ave_fin_amt, client_3m_std_fin_amt, client_3m_dr, 
                                client_3m_term, client_3m_late_days, client_3m_isLate, client_season
                                ])
    #         print("cf len:",len(client_features))

    #     print(i,(time()-s)/60)

client_features = pd.DataFrame(client_features, columns=['rownum','doc_id','client_yrs', 'client_past_inv_cnt', 
                                                     'client_annual_inv_cnt', 'client_ave_fin_amt', 
                        'client_ave_annual_fin_amt', 'client_std_fin_amt', 'client_late_days', 'client_isLate', 
                                                     'client_dr', 
                        'client_term', 'client_cty', 'client_prevyr_inv_cnt', 
                        'client_prevyr_ave_fin_amt', 'client_prevyr_std_fin_amt', 'client_prevyr_late_days', 
                        'client_prevyr_isLate', 'client_prevyr_dr', 'client_prevyr_term', 
                        'client_6m_inv_cnt', 
                        'client_6m_ave_fin_amt', 'client_6m_std_fin_amt', 'client_6m_dr', 'client_6m_term', 
                        'client_6m_late_days', 'client_6m_isLate', 
                        'client_3m_inv_cnt', 'client_3m_ave_fin_amt', 'client_3m_std_fin_amt', 'client_3m_dr', 
                        'client_3m_term', 'client_3m_late_days', 'client_3m_isLate','client_season'])
                      
# client_features.to_csv('/home/ec2-user/SCB_CODE/cam_code/data_180112/add_features_client_180113.csv')
e =time()


print('time:',(e-s)/60)
print('rowcount:',len(client_features))


client_all = client_features[['rownum','doc_id','client_yrs', 'client_past_inv_cnt', 
                                                     'client_annual_inv_cnt', 'client_ave_fin_amt', 
                        'client_ave_annual_fin_amt', 'client_std_fin_amt', 'client_late_days', 'client_isLate', 
                                                     'client_dr', 
                        'client_term', 'client_cty']]
client_prevyr = client_features[['rownum','doc_id','client_prevyr_inv_cnt', 
                        'client_prevyr_ave_fin_amt', 'client_prevyr_std_fin_amt', 'client_prevyr_late_days', 
                        'client_prevyr_isLate', 'client_prevyr_dr', 'client_prevyr_term', 'client_season'
                        ]]
client_6m = client_features[['rownum','doc_id','client_6m_inv_cnt', 
                        'client_6m_ave_fin_amt', 'client_6m_std_fin_amt', 'client_6m_dr', 'client_6m_term', 
                        'client_6m_late_days', 'client_6m_isLate']]
client_3m = client_features[['rownum','doc_id','client_3m_inv_cnt', 'client_3m_ave_fin_amt', 'client_3m_std_fin_amt', 'client_3m_dr', 
                        'client_3m_term', 'client_3m_late_days', 'client_3m_isLate','client_season']]


client_all.to_csv(data_dir+'client_all.csv', index = False)
client_prevyr.to_csv(data_dir+'client_prevyr.csv', index = False)
client_6m.to_csv(data_dir+'client_6m.csv', index = False)
client_3m.to_csv(data_dir+'client_3m.csv', index = False)
    
```
