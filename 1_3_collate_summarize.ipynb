```python
import numpy as np
import pandas as pd
from time import time
from collections import Counter
%pylab inline
```

```python
data_dir = '/home/flowcast/data/python/data_validate_180315/'

data = pd.DataFrame()
products = ['BATB', 'BDBR', 'ECIF', 'ECPR', 'IFNB', 'IFNS', 'LBDS', 'LRRP', 'PSHP', 'TRDS', 'RSRV']
print(len(data))
for product in products:
    print(product)
    filename = data_dir + 'data_3NF_180316_'+ product + '.csv'
    sub_file = pd.read_csv(filename)
    sub_file['product_'] = product
    data = data.append(sub_file, ignore_index=True)
    print(len(data))

print(data.columns)
data.describe()
```
