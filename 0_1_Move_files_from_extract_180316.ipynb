```python
def mkdir_p(path):
    try:
        os.makedirs(path)
    except OSError as exc:  # Python >2.5
        if exc.errno == errno.EEXIST and os.path.isdir(path):
            pass
        else:
            raise
```

```python
from_path = '/home/flowcast/data/FROM_SCB_TXT/OTP_data_2017_zip/'
to_path = '/home/flowcast/data/FROM_SCB_TXT/OTP_data_2017/'
```
