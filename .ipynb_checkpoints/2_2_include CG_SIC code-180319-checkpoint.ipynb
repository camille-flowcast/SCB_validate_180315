```python
#final dt_range

dt_range = []
for (x,y,z,w) in zip(data_long.doc_date, data_long.start_dt_x, data_long.end_dt, data_long.start_dt_y):
    if pd.isnull(y):
        dt_range.append(1)
    else:
        if str(x)>str(z):
            dt_range.append(0)
        else:
            if str(w)<str(x)<str(y):
                dt_range.append(0)
            else:
                dt_range.append(1)

sum(dt_range), len(dt_range)
```
