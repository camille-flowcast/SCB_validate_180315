### FOR LONG RUN (10hrs)

```python
s=time()
data=data.sort_values(by=['buyer_supplier', 'doc_date'])
data.reset_index(drop=True, inplace=True)
data['rownum']=data.index
buyer_suppliers = list(pd.unique(data.buyer_supplier))
# data.head()
sort_time = (time()-s)/60
print("sort time:", sort_time)

print(len(buyer_suppliers))

s=time()

buyer_suppliers_sample = buyer_suppliers[:20]

buyer_supplier_features = []

# for buyer_supplier in buyer_suppliers_sample:
for buyer_supplier in buyer_suppliers:
    buyer_supplier_ = data[data.buyer_supplier==buyer_supplier]
    print(buyer_supplier)
    
    for i in range(min(buyer_supplier_.rownum), max(buyer_supplier_.rownum)):
        
        if buyer_supplier_.calculate_additional_features_flag[i] == 'N':
            continue
        else:
    #         print(i)

            yesterday = buyer_supplier_.doc_date[i] - pd.DateOffset(days = 1)
            buyer_supplier_sub = buyer_supplier_[(buyer_supplier_.doc_date<=yesterday)]
    #         print(len(buyer_supplier_sub))
            try:
                buyer_supplier_yrs = (max(buyer_supplier_sub.doc_date) - min(buyer_supplier_sub.doc_date)).year +1
            except:
                buyer_supplier_yrs = 1
            buyer_supplier_past_inv_cnt = len(buyer_supplier_sub)
            buyer_supplier_annual_inv_cnt = buyer_supplier_past_inv_cnt*1./buyer_supplier_yrs*1.
            buyer_supplier_ave_fin_amt = np.mean(buyer_supplier_sub.fin_usd_amt)
            buyer_supplier_ave_annual_fin_amt = np.sum(buyer_supplier_sub.fin_usd_amt) / buyer_supplier_yrs*1.
            buyer_supplier_std_fin_amt = np.std(buyer_supplier_sub.fin_usd_amt)
            buyer_supplier_late_days = np.mean(buyer_supplier_sub.late_days)
            buyer_supplier_isLate = np.mean(buyer_supplier_sub.isLate)
            buyer_supplier_dr = np.mean(buyer_supplier_sub.dr)
            buyer_supplier_term = np.mean(buyer_supplier_sub.fin_tenor)
            buyer_supplier_cty = len(pd.unique(buyer_supplier_sub.cty_code))

    #         print(buyer_supplier_past_inv_cnt)
            prev_yr = str(buyer_supplier_.doc_date[i] - pd.DateOffset(years = 1))
            buyer_supplier_prevyr_sub = buyer_supplier_sub[
        #         (data.doc_date < data.doc_date[i]) 
                (buyer_supplier_sub.rownum < i) 
                &(buyer_supplier_sub.doc_date >=prev_yr)
        #         & (data.buyer_supplier == data.buyer_supplier[i] )
                ]
            buyer_supplier_prevyr_inv_cnt = len(buyer_supplier_prevyr_sub)
            buyer_supplier_prevyr_ave_fin_amt = np.mean(buyer_supplier_prevyr_sub.fin_usd_amt)
            buyer_supplier_season = buyer_supplier_.fin_usd_amt[i]/buyer_supplier_prevyr_ave_fin_amt
            buyer_supplier_prevyr_std_fin_amt = np.std(buyer_supplier_prevyr_sub.fin_usd_amt)
            buyer_supplier_prevyr_late_days = np.mean(buyer_supplier_prevyr_sub.late_days)
            buyer_supplier_prevyr_isLate = np.mean(buyer_supplier_prevyr_sub.isLate)
            buyer_supplier_prevyr_dr = np.mean(buyer_supplier_prevyr_sub.dr)
            buyer_supplier_prevyr_term = np.mean(buyer_supplier_prevyr_sub.fin_tenor)
        #     buyer_supplier_prevyr_cty = len(pd.unique(buyer_supplier_prevyr_sub.cty_code))


            prev_6m = str(buyer_supplier_.doc_date[i] - pd.DateOffset(months = 6))
            buyer_supplier_6m_sub = buyer_supplier_prevyr_sub[
        #         (data.doc_date < data.doc_date[i]) 
                (buyer_supplier_prevyr_sub.rownum < i) 
                &(buyer_supplier_prevyr_sub.doc_date >= prev_6m)
        #         & (data.buyer_supplier == data.buyer_supplier[i] )
                ]

            buyer_supplier_6m_inv_cnt = len(buyer_supplier_6m_sub)
            buyer_supplier_6m_ave_fin_amt = np.mean(buyer_supplier_6m_sub.fin_usd_amt)
            buyer_supplier_6m_std_fin_amt = np.std(buyer_supplier_6m_sub.fin_usd_amt)
            buyer_supplier_6m_dr = np.mean(buyer_supplier_6m_sub.dr)
            buyer_supplier_6m_term = np.mean(buyer_supplier_6m_sub.fin_tenor)
        #     buyer_supplier_6m_cty = len(pd.unique(buyer_supplier_6m_sub.cty_code))
            buyer_supplier_6m_late_days = np.mean(buyer_supplier_6m_sub.late_days)
            buyer_supplier_6m_isLate = np.mean(buyer_supplier_6m_sub.isLate)

            prev_3m = str(buyer_supplier_.doc_date[i] - pd.DateOffset(months = 3))
            buyer_supplier_3m_sub = buyer_supplier_prevyr_sub[
        #         (data.doc_date < data.doc_date[i]) 
                (buyer_supplier_prevyr_sub.rownum < i) 
                &(buyer_supplier_prevyr_sub.doc_date >= prev_3m)
        #         & (data.buyer_supplier == data.buyer_supplier[i] )
                ]

            buyer_supplier_3m_inv_cnt = len(buyer_supplier_3m_sub)
            buyer_supplier_3m_ave_fin_amt = np.mean(buyer_supplier_3m_sub.fin_usd_amt)
            buyer_supplier_3m_std_fin_amt = np.std(buyer_supplier_3m_sub.fin_usd_amt)
            buyer_supplier_3m_dr = np.mean(buyer_supplier_3m_sub.dr)
            buyer_supplier_3m_term = np.mean(buyer_supplier_3m_sub.fin_tenor)
        #     buyer_supplier_3m_cty = len(pd.unique(buyer_supplier_6m_sub.cty_code))
            buyer_supplier_3m_late_days = np.mean(buyer_supplier_3m_sub.late_days)
            buyer_supplier_3m_isLate = np.mean(buyer_supplier_3m_sub.isLate)

            buyer_supplier_features.append([i,data.doc_id[i], buyer_supplier_yrs, buyer_supplier_past_inv_cnt, buyer_supplier_annual_inv_cnt, buyer_supplier_ave_fin_amt, 
                                buyer_supplier_ave_annual_fin_amt, buyer_supplier_std_fin_amt, buyer_supplier_late_days, buyer_supplier_isLate, buyer_supplier_dr, 
                                buyer_supplier_term, buyer_supplier_cty, buyer_supplier_prevyr_inv_cnt, 
                                buyer_supplier_prevyr_ave_fin_amt, buyer_supplier_prevyr_std_fin_amt, buyer_supplier_prevyr_late_days, 
                                buyer_supplier_prevyr_isLate, buyer_supplier_prevyr_dr, buyer_supplier_prevyr_term,  
                                buyer_supplier_6m_inv_cnt, 
                                buyer_supplier_6m_ave_fin_amt, buyer_supplier_6m_std_fin_amt, buyer_supplier_6m_dr, buyer_supplier_6m_term, 
                                buyer_supplier_6m_late_days, buyer_supplier_6m_isLate, 
                                buyer_supplier_3m_inv_cnt, buyer_supplier_3m_ave_fin_amt, buyer_supplier_3m_std_fin_amt, buyer_supplier_3m_dr, 
                                buyer_supplier_3m_term, buyer_supplier_3m_late_days, buyer_supplier_3m_isLate, buyer_supplier_season
                                ])
    #         print("cf len:",len(buyer_supplier_features))

    #     print(i,(time()-s)/60)

buyer_supplier_features = pd.DataFrame(buyer_supplier_features, columns=['rownum','doc_id','buyer_supplier_yrs', 'buyer_supplier_past_inv_cnt', 
                                                     'buyer_supplier_annual_inv_cnt', 'buyer_supplier_ave_fin_amt', 
                        'buyer_supplier_ave_annual_fin_amt', 'buyer_supplier_std_fin_amt', 'buyer_supplier_late_days', 'buyer_supplier_isLate', 
                                                     'buyer_supplier_dr', 
                        'buyer_supplier_term', 'buyer_supplier_cty', 'buyer_supplier_prevyr_inv_cnt', 
                        'buyer_supplier_prevyr_ave_fin_amt', 'buyer_supplier_prevyr_std_fin_amt', 'buyer_supplier_prevyr_late_days', 
                        'buyer_supplier_prevyr_isLate', 'buyer_supplier_prevyr_dr', 'buyer_supplier_prevyr_term', 
                        'buyer_supplier_6m_inv_cnt', 
                        'buyer_supplier_6m_ave_fin_amt', 'buyer_supplier_6m_std_fin_amt', 'buyer_supplier_6m_dr', 'buyer_supplier_6m_term', 
                        'buyer_supplier_6m_late_days', 'buyer_supplier_6m_isLate', 
                        'buyer_supplier_3m_inv_cnt', 'buyer_supplier_3m_ave_fin_amt', 'buyer_supplier_3m_std_fin_amt', 'buyer_supplier_3m_dr', 
                        'buyer_supplier_3m_term', 'buyer_supplier_3m_late_days', 'buyer_supplier_3m_isLate','buyer_supplier_season'])
                      
# buyer_supplier_features.to_csv('/home/ec2-user/SCB_CODE/cam_code/data_180112/add_features_buyer_supplier_180113.csv')
e =time()


print('time:',(e-s)/60)
print('rowcount:',len(buyer_supplier_features))


buyer_supplier_all = buyer_supplier_features[['rownum','doc_id','buyer_supplier_yrs', 'buyer_supplier_past_inv_cnt', 
                                                     'buyer_supplier_annual_inv_cnt', 'buyer_supplier_ave_fin_amt', 
                        'buyer_supplier_ave_annual_fin_amt', 'buyer_supplier_std_fin_amt', 'buyer_supplier_late_days', 'buyer_supplier_isLate', 
                                                     'buyer_supplier_dr', 
                        'buyer_supplier_term', 'buyer_supplier_cty']]
buyer_supplier_prevyr = buyer_supplier_features[['rownum','doc_id','buyer_supplier_prevyr_inv_cnt', 
                        'buyer_supplier_prevyr_ave_fin_amt', 'buyer_supplier_prevyr_std_fin_amt', 'buyer_supplier_prevyr_late_days', 
                        'buyer_supplier_prevyr_isLate', 'buyer_supplier_prevyr_dr', 'buyer_supplier_prevyr_term', 'buyer_supplier_season'
                        ]]
buyer_supplier_6m = buyer_supplier_features[['rownum','doc_id','buyer_supplier_6m_inv_cnt', 
                        'buyer_supplier_6m_ave_fin_amt', 'buyer_supplier_6m_std_fin_amt', 'buyer_supplier_6m_dr', 'buyer_supplier_6m_term', 
                        'buyer_supplier_6m_late_days', 'buyer_supplier_6m_isLate']]
buyer_supplier_3m = buyer_supplier_features[['rownum','doc_id','buyer_supplier_3m_inv_cnt', 'buyer_supplier_3m_ave_fin_amt', 'buyer_supplier_3m_std_fin_amt', 'buyer_supplier_3m_dr', 
                        'buyer_supplier_3m_term', 'buyer_supplier_3m_late_days', 'buyer_supplier_3m_isLate','buyer_supplier_season']]


buyer_supplier_all.to_csv(data_dir+'buyer_supplier_all.csv')
buyer_supplier_prevyr.to_csv(data_dir+'buyer_supplier_prevyr.csv')
buyer_supplier_6m.to_csv(data_dir+'buyer_supplier_6m.csv')
buyer_supplier_3m.to_csv(data_dir+'buyer_supplier_3m.csv')
    
```
