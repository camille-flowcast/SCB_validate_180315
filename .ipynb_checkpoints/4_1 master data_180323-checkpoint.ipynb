```python
# ld_rf_preds.drop(['Unnamed: 0'],axis=1, inplace =True )
data_pred = data.merge(ld_rf_preds.drop(['Unnamed: 0','late_days_bucket'],axis=1), how = 'left', on = 'doc_id')
```

```python
# dr_rf_preds.drop(['Unnamed: 0'],axis=1, inplace =True )
data_pred = data_pred.merge(dr_rf_preds.drop(['Unnamed: 0', 'dr_bucket'],axis=1), how = 'left', on = 'doc_id')
data_pred.tail()
```

```python
# data_pred.drop(['buyer_supplier_season_x', 'client_season_x'], axis = 1, inplace = True)
data_pred.drop(['rownum'], axis = 1, inplace = True)
```

```python
from sklearn.metrics import confusion_matrix as cm
```

```python
print(cm(data_pred[data_pred.first_client_invoice==1].late_days_bucket, data_pred[data_pred.first_client_invoice==1].isLate_predict))
```

```python
industry_lookup = pd.read_csv('/home/ec2-user/SCB_DATA/sic_code_reference.csv')
industry_lookup.head()
```

```python
industry_lookup.rename(columns = {'sic_code':'industry_code'}, inplace=True)
data_pred = data_pred.merge(industry_lookup, how = 'left', on = 'industry_code')
data_pred.tail()
```

```python
data_pred['Industry'] = np.where(data_pred.industry_code==0, 'No Industry Code', data_pred.Industry)
data_pred.tail()
```

```python
len(data_pred)
```

```python
len(data)
```

```python
data_pred.drop_duplicates(inplace = True)
len(data_pred)
```

```python
Counter(data_pred.product_)
```

```python
data_pred['buyer_id'] = [int(x) for x in data_pred.buyer_id]
data_pred['cust_id'] = [y if x in ['IFNB', 'VSRP'] else z for x,y,z in zip(
    data_pred.product_, data_pred.buyer_id, data_pred.supplier_id)]
data_pred['customer_role'] = np.where(data_pred.cust_id ==data_pred.supplier_id,
                                             'SUPPLIER', 'BUYER')
data_pred['counterparty_id'] = np.where(data_pred.cust_id ==data_pred.supplier_id, 
                                               data_pred.buyer_id, data_pred.supplier_id)

data_pred[['doc_id', 'product_', 'buyer_id', 'supplier_id', 'cust_id', 
                  'customer_role', 'counterparty_id']].head(20)
```

```python
data_pred['isLate'] = np.where(data_pred.late_days>30,1,0)
Counter(data_pred.isLate)
```
