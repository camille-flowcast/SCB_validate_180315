```python
dr_coeffs = pd.read_csv(data_dir+'dr_logreg_scores_180223.csv')
dr_coeffs = dr_coeffs.transpose()
dr_coeffs['features'] = dr_coeffs.index
dr_coeffs.head()

#melt coeffs by bucket
dr_melt = dr_coeffs.melt(id_vars='features', value_name = 'dr_coeffs', var_name = 'dr_predict')
dr_melt['dr_predict'] = [int(x) for x in dr_melt.dr_predict]
dr_melt.head()
```

```python
ld_coeffs = pd.read_csv(data_dir+'LD_logreg_scores_180223.csv')
ld_coeffs = ld_coeffs.transpose()
ld_coeffs['features'] = ld_coeffs.index

ld_melt = ld_coeffs.melt(id_vars='features', value_name = 'ld_coeffs', var_name = 'isLate_predict')
ld_melt['isLate_predict'] = [int(x) for x in ld_melt.isLate_predict]
ld_melt.head()
```
